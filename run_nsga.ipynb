# Experiment number 1

```python
from metaheuristics.nsga.nsga_graph import NSGAGraphOptimizer
from metaheuristics.experiment.nsga.nsga_runner import NSGAExperimentExecuter, NSGAExperimentHyperparams

executer = NSGAExperimentExecuter(dataset_path="data/datasets.json")
params = NSGAExperimentHyperparams(
    config_id=1,
    population_size=100,
    generations=30,
    penalty_signal=0.5,
    penalty_lanes_factor=1.2
)

for dataset_id in ["1", "2", "3", "4", "5"]:
    executer.run_experiment(dataset_id=dataset_id, hyperparams=params)
```

## Summary of the experiment

```python
from metaheuristics.experiment.analyzer import ExperimentAnalyzer

for dataset_id in ["1", "2", "3", "4", "5"]:
    print(f"----------------------------------Optimization of dataset {dataset_id}--------------------------------------------")
    analyzer = ExperimentAnalyzer(f"experiments/nsga/experiment_{dataset_id}.json")
    analyzer.summarize_config(1)
```

## Hypervolume and spread

```python
runs = analyzer.get_runs_by_config_id(1)
hv_histories = analyzer.extract_metric_history(runs, "hv_history")
analyzer.plot_metric_per_iteration(hv_histories, label="HV", ylabel="Hypervolume")
```

```python
spread_histories = analyzer.extract_metric_history(runs, "spread_history")
analyzer.plot_metric_per_iteration(spread_histories, label="Spread", ylabel="Spread")
```

```python
import matplotlib.pyplot as plt

hv_final = [h[-1] for h in hv_histories]
plt.boxplot(hv_final)
plt.title("Final Hypervolume across runs")
plt.ylabel("Hypervolume")
plt.show()
```
