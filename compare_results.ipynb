```python
import json
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu
```

```python
def get_best_config_results(json_path):
    with open(json_path, 'r') as f:
        configs = json.load(f)
    best_mean_hv = -np.inf
    best_config = None
    for config in configs:
        hv_list = [run["hv_history"][-1] for run in config["runs"] if run.get("hv_history")]
        if len(hv_list) == 0:
            continue
        mean_hv = np.mean(hv_list)
        if mean_hv > best_mean_hv:
            best_mean_hv = mean_hv
            best_config = config
    hv_list = [run["hv_history"][-1] for run in best_config["runs"] if run.get("hv_history")]
    spread_list = [run["spread_history"][-1] for run in best_config["runs"] if run.get("spread_history")]
    return hv_list, spread_list, best_config["config_id"]
```

```python
n_experiments = 5

paco_dir = "experiments/paco/"
nsga_dir = "experiments/nsga/"
all_results = []
```

```python
for i in range(1, n_experiments+1):
    paco_path = f"{paco_dir}experiment_{i}.json"
    nsga_path = f"{nsga_dir}experiment_{i}.json"

    try:
        paco_hv, paco_spread, paco_best = get_best_config_results(paco_path)
        nsga_hv, nsga_spread, nsga_best = get_best_config_results(nsga_path)
    except Exception as e:
        print(f"Skipping experiment {i}: {e}")
        continue

    stat_hv, pval_hv = mannwhitneyu(paco_hv, nsga_hv, alternative="two-sided")
    stat_spread, pval_spread = mannwhitneyu(paco_spread, nsga_spread, alternative="two-sided")

    print(f"--- Experiment {i} ---")
    print(f"PACO best config: {paco_best}")
    print(f"NSGA-II best config: {nsga_best}")
    print(f"Hypervolume: PACO mean={np.mean(paco_hv):.4f}, NSGA-II mean={np.mean(nsga_hv):.4f}, Mann-Whitney p-value={pval_hv:.4g}")
    print(f"Spread: PACO mean={np.mean(paco_spread):.4f}, NSGA-II mean={np.mean(nsga_spread):.4f}, Mann-Whitney p-value={pval_spread:.4g}")

    if pval_hv < 0.05:
        print("Statistically significant difference in Hypervolume.")
    else:
        print("No significant difference in Hypervolume.")

    if pval_spread < 0.05:
        print("Statistically significant difference in Spread.")
    else:
        print("No significant difference in Spread.")

    # Save for later plotting/summary if you want
    all_results.append({
        "experiment": i,
        "paco_hv": paco_hv,
        "nsga_hv": nsga_hv,
        "paco_spread": paco_spread,
        "nsga_spread": nsga_spread,
        "paco_best": paco_best,
        "nsga_best": nsga_best,
        "pval_hv": pval_hv,
        "pval_spread": pval_spread
    })

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.boxplot([paco_hv, nsga_hv], labels=["PACO", "NSGA-II"])
    plt.title(f"Experiment {i}: Final Hypervolume")
    plt.ylabel("Hypervolume")
    plt.subplot(1, 2, 2)
    plt.boxplot([paco_spread, nsga_spread], labels=["PACO", "NSGA-II"])
    plt.title(f"Experiment {i}: Final Spread")
    plt.ylabel("Spread")
    plt.tight_layout()
    plt.show()
```
