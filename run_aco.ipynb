# Experiment 1

## Run the experiment

```python
from metaheuristics.experiment.aco.paco_runner import PACOExperimentExecuter, PACOExperimentHyperparams
from metaheuristics.experiment.analyzer import ExperimentAnalyzer

executer = PACOExperimentExecuter(dataset_path="data/datasets.json")
params = PACOExperimentHyperparams(
    config_id=1,
    n_ants=100,
    alpha=1.0,
    beta=2.0,
    rho=0.25,
    max_no_improve=25,
    penalty_signal=0.5,
    penalty_lanes_factor=1.2
)

for dataset_id in ["1", "2", "3", "4", "5"]:
    executer.run_experiment(dataset_id=dataset_id, hyperparams=params)
```

## Summary of the experiment

```python
from metaheuristics.experiment.analyzer import ExperimentAnalyzer
import numpy as np

def pad_with_nan(histories):
    max_len = max(len(h) for h in histories)
    return [h + [np.nan]*(max_len - len(h)) for h in histories]

for dataset_id in ["1", "2", "3", "4", "5"]:
    print(f"----------------------------------Optimization of dataset {dataset_id}--------------------------------------------")
    analyzer = ExperimentAnalyzer(f"experiments/paco/experiment_{dataset_id}.json")
    analyzer.summarize_config(1)
```

## Hypervolume and Spread plots

```python
runs = analyzer.get_runs_by_config_id(1)
hv_histories = analyzer.extract_metric_history(runs, "hv_history")
hv_histories_padded = pad_with_nan(hv_histories)
analyzer.plot_metric_per_iteration(hv_histories_padded, label="HV", ylabel="Hypervolume")
```

```python
spread_histories = analyzer.extract_metric_history(runs, "spread_history")
spread_histories_padded = pad_with_nan(hv_histories)
analyzer.plot_metric_per_iteration(spread_histories_padded, label="Spread", ylabel="Spread")
```

```python
import matplotlib.pyplot as plt
hv_final = [h[-1] for h in hv_histories]
plt.boxplot(hv_final)
plt.title("Final Hypervolume (HV) across runs")
plt.ylabel("HV")
plt.show()
```
